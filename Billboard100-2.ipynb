# Wrangling the Billboard top 100

## Part A

```python
import pandas as pd
import matplotlib.pyplot as plt


billboard = pd.read_csv("/Users/krummelha/Downloads/billboard.csv")
billboard.columns
billboard.drop(columns="Unnamed: 0", inplace=True)

top10songs = billboard.groupby(["song", "performer"])[["weeks_on_chart"]].max()
top10songs.sort_values(by="weeks_on_chart", ascending=False, inplace=True)
top10songs[:10]
```

The above chart shows the top songs which appeared on the Billboard Top 100 from 1958 to 2021 in descending order. According to this data, Radioactive by Imagine Dragons spent the most weeks on the chart, while Before he cheats by Carrie Underwood is the 10th most popular with 64 weeks on chart. Another interesting detail to note is that most of the songs present in this list are pop songs. 

## Part B

```python
billboard_yearfilter = billboard[billboard["year"] != 1958]
billboard_yearfilter = billboard_yearfilter[billboard_yearfilter["year"] != 2021]

billboard_yearfilter.drop_duplicates(subset=["song", "performer"], inplace=True)

billboard_yearfilter.groupby("year")[["song"]].nunique().plot(kind="line")
plt.ylabel('Number of Songs Appearing on Billboard Top 100')
plt.title('Musical Diversity from 1959 - 2020')
plt.show()
```

According to the data, it seems as though there have been large shifts in the musical diversity present within the Billboard Top 100 from 1959 - 2020. The above graph demonstrates that musical diversity peaked around the late-mid 1960s with 700 songs appearing on the Billboard. This value consistently decreased to a minimum thirty years thereafter in the early 2000s, with just above 300 songs appearing on the billboard that year. However, there is a local peak in 2010 with around 475 songs, and a sharp increase in songs appearing on the Billboard shortly thereafter, indicating that musical diversity is again on the rise according to the Billboard Top 100. 

## Part C

```python
billboard_10weekhit = billboard[billboard["weeks_on_chart"] >= 10]
billboard_10weekhit.drop_duplicates(subset=["performer", "song"], inplace=True)

top19 = billboard_10weekhit["performer"].value_counts()

top19[top19 >= 30].plot(kind="bar", title='19 Artists with at least 30 Ten Week Hits')
plt.ylabel('Number of 10-Week Hits')
plt.title('Artists with at least 30 Ten Week Hits')
plt.show()
```

The above graph lists 19 artists who have had at least 30, 10-Week Hits on the Billboard Top 100 from 1959 - 2020. While artists from a  diverse amount of genres appear on this list, country and pop artists are the most represented, with Drake being the sole rap artist. Elton John takes the top spot with 8 more 10-Week hits than Madonna, and three country artists tail behind. Moreover, four artists on the list, Brad Paisley, George Straight, Jason Aldean, and Neil Diamond all share the spotlight for 30, 10-Week Hits and interestingly, are all country artists, besides Neil Diamond. Utimately, while the Billboard Top 100 has been mostly comprised of songs from country artists, pop songs by Elton John are the highest played.
